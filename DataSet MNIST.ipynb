```python
import pandas as pd
import numpy as np

AllData = pd.read_csv('E:\\Machine Learning\\Dataset MNIST\\train.csv')
label =AllData.label
AllData.drop(['label'], axis = 1, inplace = True)
AllData[AllData != 0] = 1
```

```python
matrix = []
for k in range(len(AllData)):
    p = np.array(AllData.loc[k]) 
    BL = []
    for i in range(0, len(p), 28):
        mat = []
        mat = p[i: i+28]
        BL.append(mat) 
    matrix.append(BL)  
```

```python
#pca filter study
"which is used to remove unwanted area from a collection of images try using matplot lib filter then opencv"

import matplotlib.pyplot as plt
for r in range( 100):
    plt.imshow(matrix[r], cmap=plt.get_cmap('gray'))
    plt.show()
```
